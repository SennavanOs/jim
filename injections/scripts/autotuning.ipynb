```python
def compute_gamma(acceptance_rate, target_rate=0.574, max_adjustment=0.1, scale_factor=0.1):
    deviation = acceptance_rate - target_rate
    deviation = np.minimum(max_adjustment, scale_factor * deviation)
    gamma = np.exp(deviation)
    return gamma

def compute_gamma(acceptance_rate, target_rate=0.574, max_adjustment=0.1, scale_factor=0.1):
    # deviation = acceptance_rate - target_rate
    # deviation = np.minimum(max_adjustment, scale_factor * deviation)
    # gamma = np.exp(deviation)
    
    return 0.2 * (acceptance_rate - 0.574) + 1

# Visualize this
accs = np.linspace(0, 1, 100)
gamma = compute_gamma(accs)

print(compute_gamma(0.574))


plt.plot(accs, gamma)
plt.xlabel("Acceptance")
plt.ylabel("Gamma")
plt.show()
```
